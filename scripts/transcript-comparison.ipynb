### Paths

```python
PROJECT_ROOT = Path("__file__").resolve().parent.parent
ANALYSIS_DIR = PROJECT_ROOT / "samples" / "zurich_subset"
WHISPER_DIT_SMALL = ANALYSIS_DIR / "whisper" / "dialect-ignorant-transcript-small.tsv"
WHISPER_DIT_MEDIUM = ANALYSIS_DIR / "whisper" / "dialect-ignorant-transcript-medium.tsv"
WHISPER_DIT_LARGE_V1 = ANALYSIS_DIR / "whisper" / "dialect-ignorant-transcript-large-v1.tsv"

WHISPER_DAT_LARGE_V2 = ANALYSIS_DIR / "whisper" / "dialect-aware-transcript-large-v2.tsv"
WHISPER_DAT_LARGE_V3 = ANALYSIS_DIR / "whisper" / "dialect-aware-transcript-large-v3.tsv"
FHNW_DAT_LARGE = ANALYSIS_DIR / "fhnw" / "dialect-aware-transcript-fhnw.tsv"
```

### Merge transcripts to compare

```python
df_dit_small = pd.read_csv(WHISPER_DIT_SMALL, sep="\t", encoding="utf-8-sig")
df_dit_medium = pd.read_csv(WHISPER_DIT_MEDIUM, sep="\t", encoding="utf-8-sig")
df_dit_large_v1 = pd.read_csv(WHISPER_DIT_LARGE_V1, sep="\t", encoding="utf-8-sig")
df_dat_large_v2 = pd.read_csv(WHISPER_DAT_LARGE_V2, sep="\t", encoding="utf-8-sig")
df_dat_large_v3 = pd.read_csv(WHISPER_DAT_LARGE_V3, sep="\t", encoding="utf-8-sig")
df_dat_fhnw = pd.read_csv(FHNW_DAT_LARGE, sep="\t", encoding="utf-8-sig")

from functools import reduce

frames = [
    df_dat_large_v2[["path", "sentence", "DAT"]].rename(columns={"DAT": "DAT_v2"}),
    df_dat_large_v3[["path", "DAT"]].rename(columns={"DAT": "DAT_v3"}),
    df_dat_fhnw[["path", "DAT"]].rename(columns={"DAT": "DAT_fhnw"}),
    df_dit_small[["path", "DIT"]].rename(columns={"DIT": "DIT_small"}),
    df_dit_medium[["path", "DIT"]].rename(columns={"DIT": "DIT_medium"}),
    df_dit_large_v1[["path", "DIT"]].rename(columns={"DIT": "DIT_large_v1"}),
]

df = reduce(lambda left, right: pd.merge(left, right, on="path", how="inner"), frames)

print(f"Clips: {len(df)}")
df.head()
```
