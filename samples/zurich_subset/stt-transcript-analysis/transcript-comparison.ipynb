## Transcript Comparison
Comparison of multiple Transcripts (Whisper & FHNW Model) to determine DIT and DAT material.

### Importts

```python
import sys
import importlib
import pandas as pd
from pathlib import Path

SCRIPTS_DIR = Path.cwd() / "scripts" if (Path.cwd() / "scripts").exists() else Path.cwd()
sys.path.insert(0, str(SCRIPTS_DIR))
PROJECT_ROOT = SCRIPTS_DIR.parent
```

### Paths

```python
PROJECT_ROOT = Path("__file__").resolve().parent.parent.parent.parent
ANALYSIS_DIR = PROJECT_ROOT / "samples" / "zurich_subset" / "stt-transcript-analysis"

WHISPER_SMALL = ANALYSIS_DIR / "whisper" / "whisper-small.tsv"
WHISPER_MEDIUM = ANALYSIS_DIR / "whisper" / "whisper-medium.tsv"
WHISPER_LARGE_V1 = ANALYSIS_DIR / "whisper" / "whisper-large-v1.tsv"
WHISPER_LARGE_V2 = ANALYSIS_DIR / "whisper" / "whisper-large-v2.tsv"
WHISPER_LARGE_V3 = ANALYSIS_DIR / "whisper" / "whisper-large-v3.tsv"
WHISPER_LARGE_V1_PROMPTED = ANALYSIS_DIR / "whisper" / "whisper-large-v1-prompted.tsv"
WHISPER_LARGE_V2_PROMPTED = ANALYSIS_DIR / "whisper" / "whisper-large-v2-prompted.tsv"
WHISPER_LARGE_V3_PROMPTED = ANALYSIS_DIR / "whisper" / "whisper-large-v3-prompted.tsv"
FHNW_LARGE = ANALYSIS_DIR / "fhnw" / "fhnw-large.tsv"
```

### Merge transcripts to compare

```python
df_whisper_small = pd.read_csv(WHISPER_SMALL, sep="\t", encoding="utf-8-sig")
df_whisper_medium = pd.read_csv(WHISPER_MEDIUM, sep="\t", encoding="utf-8-sig")
df_whisper_large_v1 = pd.read_csv(WHISPER_LARGE_V1, sep="\t", encoding="utf-8-sig")
df_whisper_large_v2 = pd.read_csv(WHISPER_LARGE_V2, sep="\t", encoding="utf-8-sig")
df_whisper_large_v3 = pd.read_csv(WHISPER_LARGE_V3, sep="\t", encoding="utf-8-sig")
df_whisper_fhnw = pd.read_csv(FHNW_LARGE, sep="\t", encoding="utf-8-sig")

from functools import reduce

frames = [
    df_whisper_fhnw[["path", "sentence", "transcript"]].rename(columns={"transcript": "transcript_fhnw"}),
    df_whisper_large_v1[["path", "transcript"]].rename(columns={"transcript": "transcript_v1"}),
    df_whisper_large_v2[["path", "transcript"]].rename(columns={"transcript": "transcript_v2"}),
    df_whisper_large_v3[["path", "transcript"]].rename(columns={"transcript": "transcript_v3"}),
    df_whisper_medium[["path", "transcript"]].rename(columns={"transcript": "transcript_medium"}),
    df_whisper_small[["path", "transcript"]].rename(columns={"transcript": "transcript_small"}),
]

df_transcripts = reduce(lambda left, right: pd.merge(left, right, on="path", how="inner"), frames)
df_transcripts.head()
```

### Prompted

```python
df_whisper_large_v1_prompted = pd.read_csv(WHISPER_LARGE_V1_PROMPTED, sep="\t", encoding="utf-8-sig")
df_wishper_large_v2_prompted = pd.read_csv(WHISPER_LARGE_V2_PROMPTED, sep="\t", encoding="utf-8-sig")
df_whisper_large_v3_prompted = pd.read_csv(WHISPER_LARGE_V3_PROMPTED, sep="\t", encoding="utf-8-sig")

from functools import reduce

frames = [
    df_whisper_large_v1_prompted[["path", "sentence", "transcript"]].rename(columns={"transcript": "transcript_v1_prompted"}),
    df_wishper_large_v2_prompted[["path", "transcript"]].rename(columns={"transcript": "transcript_v2_prompted"}),
    df_whisper_large_v3_prompted[["path", "transcript"]].rename(columns={"transcript": "transcript_v3_propmpted"}),
]

df_prompted = reduce(lambda left, right: pd.merge(left, right, on="path", how="inner"), frames)
df_prompted.head()
```
